# Import libs

```python
import pandas as pd

import time

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

# Functions

```python
def save_log():
    """saves the position and value of the cells into a dic"""
    
    tiles_selector = ('body > div > div > section.elementor-section.elementor-top-section.elementor-element.elementor-element-00e5059.'
            'elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div > div'
            ' > div > div > div > div.game-container > div.tile-container > div.tile.tile')
    
    #waits for the page to load
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, tiles_selector)))
    
    # gets the class name of the tiles
    tiles = browser.find_elements_by_css_selector(tiles_selector)

    #tranform the WebElement in a dict with the position (key) and number (value) for each tile
    current_cells = dict()
    for tile in tiles:
        #saves the table 
        current_cells[tile.get_attribute('class').split("ion-")[1][:3]] = tile.text

    return current_cells

def move_cells(direction):
    """make a move to the direction in the arg"""
    browser.find_element_by_tag_name('html').send_keys(direction)
    
def game_over():
    #creating a var with the game-over/try-again button
    game_over_button = browser.find_element_by_class_name('retry-button')
    return True if game_over_button.is_displayed() else False
    
def play_again():
    # restart the game if its already over
    game_over_button = browser.find_element_by_class_name('retry-button')
    if game_over_button.is_displayed():
        print('restarting')
        game_over_button.click()
        

def play_game():
    """plays the game until game over and returns a dic where each item a dic with k,v = {cell position: cell value }"""
    
    #start a timer
    start_time = time.time()
    #create and empty list to receive the dics with the table 
    play_data = {}
    #zero the counter
    c = 0
    #loop through the moves and save the table of each move
    #while game_over() == False:
    for x in range(2):
        for z in range(5):    

            move_cells(move_right)
            play_data[c] = save_log()
            c += 1
            print(c)
            move_cells(move_up)
            play_data[c] = save_log()
            c += 1
            
        move_cells(move_right)
        play_data[c] = save_log()
        c += 1
        move_cells(move_left)
        play_data[c] = save_log()
        c += 1 
    
    #get the score    
    score = browser.find_element_by_class_name('score-container').text
    print (f"It took me {round(time.time() - start_time)} seconds to make {score} points!")
    
    return play_data
    
move_up = Keys.UP
move_down = Keys.DOWN
move_right = Keys.RIGHT
move_left = Keys.LEFT 
```

# Open webdriver

```python
#Open webdriver to site
browser = webdriver.Chrome()

url = 'https://play2048.com'
browser.get(url)
```

# tests
